## Machine Learning

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")
```

```python
df = pd.read_csv("Superstore.csv")
df.head()
```

## Question 1
Load the dataset and display the first five rows.

```python
df.head()
```

## Question 2
Calculate total sales per region (sorted descending).

```python
df.groupby("Region")["Sales"].sum().sort_values(ascending=False)
```

## Question 3
Countplot for product categories.

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="Category")
plt.title("Distribution of Orders by Category")
plt.show()
```

The countplot shows the distribution of orders across different product categories. It helps identify which category has the highest number of orders.

## Question 4
Boxplot of profit across regions.

```python
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x="Region", y="Profit")
plt.title("Profit Distribution by Region")
plt.show()
```

The boxplot shows how profit varies across regions. Some regions have higher median profit and more variability than others.

## Question 5
Total profit per category (descending order).

```python
df.groupby("Category")["Profit"].sum().sort_values(ascending=False)
```

## Question 6
Relationship between discount and profit.

```python
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x="Discount", y="Profit")
plt.title("Discount vs Profit")
plt.show()
```

The scatterplot shows a negative relationship between discount and profit. Higher discounts are often associated with lower or negative profit.
